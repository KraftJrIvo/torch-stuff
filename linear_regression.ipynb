```python
import torch
import numpy as np
import math
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline

import random

def dataset(n=100):
    xs = []
    ys = []
    for i in range(n):
        x = i * 4 / n
        xs.append(x)
        #ys.append(math.sin(x))
        ys.append(x * 0.666 + 0.05 * (random.gauss() * 2 - 1))
        #ys.append(abs(x - 2))
    return torch.tensor(xs), torch.tensor(ys)

def plot(once, fig, hdisplay, ax1, ax2, xs_train, ys_train, xs_test, ys_test, epochs, train_losses, test_losses):    
    ax1.clear()
    ax2.clear()
    if once:
        ax1.set_xlabel('x')
        ax1.set_ylabel('y')
        ax1.set_title('inference')
        ax1.title.set_fontsize(9)
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Loss')
        ax2.set_title('Training Loss')
        ax1.scatter(xs_train, ys_train, c='g', label='Training Data')
        ax1.scatter(xs_test, ys_test, c='r', label='Result')
        ax2.plot(epochs, train_losses, c='g', label='Training Loss')
        ax2.plot(epochs, test_losses, c='r', label='Test Loss')
    else:
        ax1.scatter(xs_train, ys_train, c='g', label='Training Data')
        ax1.scatter(xs_test, ys_test, c='r', label='Result')
        ax2.plot(epochs, train_losses, c='g', label='Training Loss')
        ax2.plot(epochs, test_losses, c='r', label='Test Loss')
    hdisplay.update(fig)

class LinearRegressionModel(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.k = torch.nn.Parameter(torch.rand(1, requires_grad=True))
        self.b = torch.nn.Parameter(torch.rand(1, requires_grad=True))
        
    def forward(self, x):
        return x * self.k + self.b

# creating dataset
xs_train, ys_train = dataset(100)
xs_test, ys_test = dataset(23)

# creating model
random.seed(42)
torch.manual_seed(42)
model = LinearRegressionModel()

# creating optimizer
optimizer = torch.optim.Adam(params = model.parameters())

# for plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
hdisplay = display("", display_id=True)

# train preparation
loss_fn = torch.nn.L1Loss()
EPOCHS = 2000
epochs = []
train_losses = []
test_losses = []
for epoch in range(EPOCHS):
    # train step
    model.train()
    ys_pred = model(xs_train)
    loss = loss_fn(ys_pred, ys_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # test evaluation
    model.eval()
    with torch.inference_mode():
        test_pred = model(xs_test)
        test_loss = loss_fn(test_pred, ys_test)
        # for plotting
        if epoch % (EPOCHS / 50) == 0:
            epochs.append(epoch)
            train_losses.append(loss.item())
            test_losses.append(test_loss.item())
            plot(epoch == 0, fig, hdisplay, ax1, ax2, xs_train, ys_train, xs_test, test_pred, epochs, train_losses, test_losses)
            print(f'Epoch: {epoch} | train loss: {round(loss.item(), 6)} | test loss: {round(test_loss.item(), 6)}')
```
